```python
#!/usr/bin/env python
""" Grafica-E1-E2-Diferencia_v3.0 Genera una grafica para comparar los valores de E1, E2 Y Diferencia."""

__author__  = 'Paola Cornejo' + 'Sarai Reyes' 

import re
import math
import pylab as p
import numpy as np

#SUBRUTINAS
def AverageDictionary (mydict):
    """#Obtiene el promedio de los valores acordes al la llave que se encuentra en el diccionario donde se gurdan"""
    count=range(0,len(mydict),)
    average=[]
    for i in count:
        values=[float(j.split(' ')[i]) for j in mydict.values()]
        average.append(np.mean(values))
    return (average)

def SequenceValueDictionary (filex):
    """Lee los archivos y los tranforma a diccionarios"""
    data=filex.readlines() 
    lseq=data[1::2]
    lval=data[2::2]
    dict1= {key:value for key, value in zip(lseq,lval)}
    average= AverageDictionary(dict1)
    return (average, {key:value for key, value in zip(lseq,average)})

#Abre los archivos
path= 'C:/Users/Toshiba/Dropbox/LCG/4to semestre/Bioinfo/proyectos/Estrcutural/Tarea 2'
file1=open (path + '/VALORES DE E1.txt', 'r')
file2=open (path + '/VALORES DE E2.txt', 'r')
file3=open (path + '/VALORES DE D.txt', 'r')

#Llama a las subrutinas para obtener una lista de promedios y el diccionario
average1, dic1= SequenceValueDictionary(file1)
average2, dic2= SequenceValueDictionary(file2)
average3, dic3= SequenceValueDictionary(file3)

#GRAFICA
n = len(dic1)
X = np.arange(n)

#Define parametros de la grafica
ax=p.subplot()
ax.plot(X, average1, linestyle=':', color='b', label = "Valores de E1")
ax.set_ylim(-20,3)
ax.plot(X, average2, linestyle='-', color='g', label = "Valores de E2")
ax.plot(X, average3, linestyle='--', color='r', label = "Valores de Diferencia")

legend=p.legend(loc=1)
ax.legend(loc=5)
p.xlabel('Secuencias')
p.ylabel('Valores')
p.title('Grafica de valores E1, E2 Y D')

p.show()
```
